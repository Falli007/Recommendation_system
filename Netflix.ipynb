```python
netflix=netflix[['id','genre','title','overview']]
netflix
```

```python
netflix['tags']=netflix['overview']+netflix['genre']
```

```python
netflix
```

```python
new_data = netflix.drop(columns=['genre', 'overview'])
```

```python
new_data
```

1.Bag of word 2. TFID


```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['title'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommended(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommended('Your Eyes Tell')
```
